# Importing packages

```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading dataset

```python
df=sns.load_dataset('iris')
df.head(10)
```

# Information about dataset

```python
df.info()
```

```python
df.describe()
```

# Training Model

```python
classifier= tree.DecisionTreeClassifier()
```

```python
classifier.fit(df.drop("species",axis=1),df.species)
```

# Predicting Output

```python
classifier.predict([[2,2,2,2],[1,8,8,5]])
```

# Visualization of Tree

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(classifier)
```
